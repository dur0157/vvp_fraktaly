Vykreslovanie frakt√°lov pomocou matplotlib.plt

```python
from mandelbrot import Mandelbrot
from julia import Julia
from visualization import FractalVisualizer

# Parametre pre Mandelbrotovu mno≈æinu
xmin, xmax, ymin, ymax = -2.0, 1.0, -1.5, 1.5
width, height = 800, 800
max_iter = 256

# Inicializ√°cia a generovanie Mandelbrotovej mno≈æiny
mandelbrot = Mandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter)
mandelbrot_set = mandelbrot.generate()

# Parametre pre Juliovu mno≈æinu
c = complex(-0.7, 0.27015)
julia_params = {
    'xmin': -1.5, 'xmax': 1.5,
    'ymin': -1.5, 'ymax': 1.5,
    'width': 800, 'height': 800,
    'c': c, 'max_iter': max_iter
}

# Inicializ√°cia a generovanie Juliovej mno≈æiny
julia = Julia(**julia_params)
julia_set = julia.generate()

# Inicializ√°cia vizualiz√©ra
visualizer = FractalVisualizer(xmin, xmax, ymin, ymax, width, height, max_iter)

# Vizualiz√°cia Mandelbrotovej mno≈æiny
visualizer.show_fractal(mandelbrot_set, 'Mandelbrotova mno≈æina', 'inferno')

# Vizualiz√°cia Juliovej mno≈æiny
visualizer.show_fractal(julia_set, f'Juliova mno≈æina pre c = {julia_params["c"]}', 'inferno')
```
